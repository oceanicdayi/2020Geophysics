```python
import os
```

```python
def create_GMT_bat(
    symbol_color = "blue",
    wet_color = "white",
    dry_color = "100/100/0",
    river_level = "a",
    output_fname = "GMT_2",
    output_format = "jpg",
    symbol = "a",
    rrange = "119/123/21/26"):
    
    GMT_cmds = []
    GMT_cmds.append(f"gmt begin {output_fname} {output_format} \n")
    GMT_cmds.append(f"gmt coast -R{rrange} -JM10 -B1 -Df -G{dry_color} -S{wet_color} -I{river_level}\n")
    GMT_cmds.append(f"gmt plot data.txt -S{symbol}0.5 -G{symbol_color} \n")
    GMT_cmds.append(f"gmt end show \n")

    with open("gmt_run.bat","w") as fp:
        for i in GMT_cmds:
            fp.write(i)
```

```python
create_GMT_bat()

while True:
    print("="*30)
    print("Program for plotting earthquakes")
    print("="*30)
    print("1) Enter epicenters ")
    print("2) Set symbols  ")
    print("3) Set color for symbol")
    print("4) Set range ")
    print("5) Plot figure ")
    print("6) Show data")
    print("7) Show GMT file")
    print("8) Set color for wet area")
    print("9) Set color for dry area")
    print("10) Set river level")
    print("11) Set output filename")
    print("12) Set output format")
    print("0) quit ")
    print("="*30)
    ans = input("Enter your choice :　")
    ans = int(ans)
    
    
    
    if ans==0:
        print("End program")
        break
    elif ans == 1:
        
        n_data = input("How many data would you like to enter ? ")
        n_data = int(n_data)

        data=[]

        with open ("data.txt","w") as fp:
            for i in range(n_data):
                lon = input("Please enter longitude: ")
                lat = input("Please enter latitude: ")
                ss = lon + " " + lat + "\n"
                print(ss)
                data.append(ss)
        
                fp.write(ss)
    elif ans == 2:
        aa=input("Enter symbol: ")
        create_GMT_bat(symbol=aa)
        
    elif ans == 3:
        symbol_color=input("Enter color: ")
        create_GMT_bat(symbol_color=symbol_color)

        

            
    elif ans == 5:
        os.system("gmt_run.bat")
            
    elif ans == 6:
        with open("data.txt","r") as fp:
            dd=fp.readlines()
        for i in dd:
            print(i)
            
    elif ans == 7:
        with open("gmt_run.bat","r") as fp:
            dd=fp.readlines()
        for i in dd:
            print(i)
    else:
        print("Wrong input !!!")
```
